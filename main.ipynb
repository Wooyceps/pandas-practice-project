# Changing the data

```python
df_csv['Total'] = df_csv.iloc[:, 4:10].sum(axis=1)
df_csv = df_csv.sort_values(['Total'], ascending=False)
```

```python
cols = list(df_csv.columns.values)

df_csv = df_csv[cols[0:4] + [cols[-1]] + cols[4:12]]
df_csv.head(5)
```

```python
df_csv.to_csv('modified.csv', index=False)
df_csv.to_excel('modified.xlsx', index=False)
df_csv.to_csv('modified.txt', index=False, sep='\t')
```
